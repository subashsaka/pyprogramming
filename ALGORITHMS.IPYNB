SEARCHING ALGORITHMS

-LINEAR SEARCH-

```python
def linearsearch(arr,target):
    for i in range(len(arr)):
        if arr[i]==target:
            return 1
    return -1

arr = [10, 20, 30, 40, 50]
print(linearsearch(arr,30))
```

-BINARY SEARCH-

```python
def binarysearch(arr,target):
    low,high = 0,len(arr)-1
    while low<=high:
        mid = (low+high)//2
        if arr[mid] == target:
            return 1
        elif arr[mid]<target:
            low = mid+1
        else:
            high = mid - 1
    return -1
arr = [10, 20, 30, 40, 50]
print(binarysearch(arr, 30))
```
